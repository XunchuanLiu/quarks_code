```python
def plot_source(fig,showsource,xrange,yrange,rest_value=220.7090165*u.GHz,threshold=18*u.mJy/u.beam,
               vl=-4,vr=4):
    TM1menbers = [
    "member.uid___A001_X1590_X341e",
    "member.uid___A001_X1590_X3417",
    "member.uid___A001_X1590_X33d1",
    "member.uid___A001_X1590_X33fb",
    "member.uid___A001_X1590_X3402",
    "member.uid___A001_X1590_X3410",
    "member.uid___A001_X1590_X3425",
    "member.uid___A001_X1590_X342c",
    "member.uid___A001_X1590_X3433",
    ]

    T=t.Table.read('sourcetable.csv')
    Tsources = np.array(list(T['source']))

    for m in TM1menbers:
        files=glob.glob(
                '../../2021.1.00095.S/science_goal.uid___A001_X1590_X33cf/*/%s/*/*_sci.spw27.cube.I.pbcor.fits' %m)
        for f in files:
            s = (f.split('/')[-1].split('.')[2][:-4])
            s=s.replace('p','+')
            if s!=showsource:
                continue
            assert s in Tsources
            dex = np.argmax(Tsources == s)
            vlsr = T['vlsrnew'][dex]
            cube = sc.SpectralCube.read(f).with_spectral_unit('km/s',velocity_convention='radio',
                                                             rest_value=rest_value)
            CH3CNcube = cube.spectral_slab( (vlsr-6)*u.km/u.s, (vlsr+6)*u.km/u.s )
            CH3CNcube_masked = CH3CNcube.with_mask(CH3CNcube > threshold)
            m0 = CH3CNcube_masked.moment0()
            m1 = CH3CNcube_masked.moment1()
            plt.figure(fig)
            plt.subplot(121, projection=m0.wcs)
            plt.imshow(m0.array,origin='lower')
            plt.xlim(xrange)
            plt.ylim(yrange)
            plt.text(0.1,0.9,s,transform=plt.gca().transAxes)
            plt.colorbar()
            plt.subplot(122, projection=m0.wcs)
            plt.imshow(m1.array,origin='lower',cmap='Spectral_r',vmin=vlsr+vl,vmax=vlsr+vr)
            plt.xlim(xrange)
            plt.ylim(yrange)
            plt.colorbar()   
```

```python
fig=plt.figure(figsize=(9,4))
s = 'I18434-0242'
plot_source(fig,s,(490,610),(295,415))
plt.savefig('figure/%s.pdf' %s, bbox_inches='tight')
```
